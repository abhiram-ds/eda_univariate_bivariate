## <font color=blue>Data Cleaning for Application Dataframe</font>

### <font color=blue> Start the analysis of Continuos variables (NUMBER fields) </font>

```python
appl_df[['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE']].describe()
```

```python
#CREATE DERIVED COLUMNS FROM THE AMOUNT VARIABLES 
appl_df['CREDIT_TO_INCOME_RATIO'] = round(appl_df['AMT_CREDIT']/appl_df['AMT_INCOME_TOTAL'],2)
appl_df['ANNUITY_TO_INCOME_RATIO'] = round(appl_df['AMT_ANNUITY']/appl_df['AMT_INCOME_TOTAL'],2)
```

```python
appl_df[['CREDIT_TO_INCOME_RATIO','ANNUITY_TO_INCOME_RATIO']].describe()
```

```python
#CHECK THE COLUMNS WITH DAYS VALUES
appl_df[['DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']].describe()
```

```python
#For DAYS, entire data is given with negative sign, probably the calculation was done with ACTUAL DATE - Application Date
#So convert the value to positive and compute the age in terms of years

appl_df['AGE_IN_YEARS'] = round(abs(appl_df['DAYS_BIRTH'])/(365),2)
appl_df['DAYS_BIRTH'] = abs(appl_df['DAYS_BIRTH'])
appl_df['DAYS_EMPLOYED'] = abs(appl_df['DAYS_EMPLOYED'])
appl_df['DAYS_REGISTRATION'] = abs(appl_df['DAYS_REGISTRATION'])
appl_df['DAYS_ID_PUBLISH'] = abs(appl_df['DAYS_ID_PUBLISH'])
appl_df['DAYS_LAST_PHONE_CHANGE'] = abs(appl_df['DAYS_LAST_PHONE_CHANGE'])
```

```python
# Check the Fields after update
appl_df[['AGE_IN_YEARS','DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH','DAYS_LAST_PHONE_CHANGE']].describe()
```

#### <font color=blue> Impute the high OUTLIER value in DAYS_EMPLOYED column </font>

```python
##CHECK THE COUNT OF APPLICATIONS BASED ON TARGET VALUE WITH HIGH OUTLIER  VALUE OF DAYS_EMPLOYED
appl_df.loc[appl_df['DAYS_EMPLOYED'] == 365243]['TARGET'].value_counts()/len(appl_df.loc[appl_df['DAYS_EMPLOYED'] == 365243].index) * 100
```

#### <font color=blue> Analysisng the data, we see that the abnormal outlier value doesn't really mean that the data is corrupted. Only this 1 columns needs to be imputed </font>

```python
appl_df['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace=True)
```

#### <font color=blue> Check the data after imputing </font>

```python
appl_df[['AGE_IN_YEARS','DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']].describe()
```

## <font color=blue> START THE ANALYSIS TO FIND CORRELATION BETWEEN VARIABLES WITH THE TARGET COLUMNS </font>

```python
#Need to find correlation  with Target to draw univariate, bivariate and segmented univariate
corr_df = appl_df.corr()['TARGET'].sort_values()
print('Positive Correlations: ', corr_df.tail(10))
print('Negative Correlations: ', corr_df.head(10))
```

### <font color=blue> The correlation is on both sides i.e. few variables affecting the loan default (TARGET=1) behavior positively as well as some variables impacting negative. Negative impact implies less loan default chance and hence indicator of credit worthy applicant. We will be considering the top variables on either sides with > 0.05 (absolute) correlation. </font>

```python
print('Driver Variables Correlations: ', corr_df[abs(corr_df.values) >= 0.05])
```

```python
##GET THE DATA DESCRIPTION OF ALL THE CORRELATED COLUMNS
appl_df[['EXT_SOURCE_3','EXT_SOURCE_2','AGE_IN_YEARS','DAYS_BIRTH','DAYS_EMPLOYED','DAYS_LAST_PHONE_CHANGE','GENDER_INDICATOR','DAYS_ID_PUBLISH','REG_CITY_NOT_WORK_CITY','REGION_RATING_CLIENT','REGION_RATING_CLIENT_W_CITY']].describe()
```

### <font color = blue> Here, of all numerical vakues, AGE_IN_YEARS is closer to TARGET which is 0.078239. Hence we Will consider this for univariate, bivariate and segmented univariate with Target
<font color = blue>Age Univariate graph </font>


```python
plt.figure(figsize = (10, 5))
plt.hist(appl_df['AGE_IN_YEARS'] , edgecolor = 'k', bins = 25)
plt.title('AGE UNIVARIATE ANALYSIS')
plt.xlabel('AGE (IN YEARS)')
plt.ylabel('APPLICATION COUNT')
plt.show()
```

 ### <font color = blue> To further understand the data distribution across age, we will do a Kernel density estimate / Segmented Univariate analysis. </font>

```python
plt.figure(figsize = (8, 5))
sns.kdeplot(appl_df.loc[appl_df['TARGET'] == 0, 'AGE_IN_YEARS'] , label='Target = 0')
sns.kdeplot(appl_df.loc[appl_df['TARGET'] == 1, 'AGE_IN_YEARS'] , label='Target = 1')
plt.title('AGE_IN_YEARS SEGMENTED UNIVARIATE');
plt.xlabel('AGE_IN_YEARS WITH TARGET')
plt.show()
```

### <font color=blue> From the above PLOT,we can understand that the high default TARGET=1 is around the age group of 27-32. While safe customers are those with Age > 32 </font>

### <font color = blue> Let us further analyse this with BI-VARIATE CATEGORICAL DATA </font>

```python
plt.figure(figsize = (10, 5))
sns.kdeplot(appl_df.loc[(appl_df['CODE_GENDER'] == 'F') & (appl_df['TARGET'] == 0), 'AGE_IN_YEARS'] , label='FEMALE-0')
sns.kdeplot(appl_df.loc[(appl_df['CODE_GENDER'] == 'F') & (appl_df['TARGET'] == 1), 'AGE_IN_YEARS'] , label='FEMALE-1')
sns.kdeplot(appl_df.loc[(appl_df['CODE_GENDER'] == 'M') & (appl_df['TARGET'] == 0), 'AGE_IN_YEARS'] , label='MALE-0')
sns.kdeplot(appl_df.loc[(appl_df['CODE_GENDER'] == 'M') & (appl_df['TARGET'] == 1), 'AGE_IN_YEARS'] , label='MALE-1')
plt.title('AGE_IN_YEARS & GENDER BI-VARIATE CATEGORICAL');
plt.xlabel('AGE_IN_YEARS & GENDER WITH TARGET')
plt.show()
```

### <font color=blue> From the above BI-VARIATE CATEGORICAL kde plot,we can infer that MALE & FEMALE APPLICANTS within the age group of 27-32 are high risk, where as  that the reliable customer age group starts around 33 for MALE applicats and around 35 for FEMALE applicants.</font>

### <font color=blue> Merge the Current and Previous application Data Sets </font>

```python
##MERGE THE PREVIOUS APPLICATION DATA INTO THE CURRENT APPLICATION DATAFRAME
merged_appl_df = appl_df.merge(prev_appl_df, on='SK_ID_CURR', how='left')
print("Merged applications has {0} rows and {1} columns".format(merged_appl_df.shape[0], merged_appl_df.shape[1]))
```

```python
##CHECK THE STATUS OF APPLICATIONS IN MERGED DATA SET
merged_appl_df['NAME_CONTRACT_STATUS'].value_counts()
```

```python
##CHECK THE APPROVED APPLICATION DATA
merged_appl_df.loc[merged_appl_df['NAME_CONTRACT_STATUS']=='Approved'].head()
```

```python
##PLOT THE EMPLOYMENT EXPERIENCE PLOT BASED ON THE APPROVED STATUS OF APPLICATION FOR MERGED DATA
merged_appl_emp = merged_appl_df.loc[merged_appl_df['NAME_CONTRACT_STATUS']=='Approved'][['DAYS_EMPLOYED','TARGET']]
merged_appl_emp['EMP_YEARS'] = merged_appl_emp['DAYS_EMPLOYED'] / 365
merged_appl_emp['EMP_YEARS_BINS'] = pd.cut(merged_appl_emp['EMP_YEARS'], bins = np.linspace( 0.00, 50.00, num = 20))
merged_appl_emp_ctgy  = merged_appl_emp.groupby('EMP_YEARS_BINS').mean()
print(merged_appl_emp_ctgy)
```

```python
##PLOT THE HISOGRAM WITH EMP EXPERIENCE BINS AND ANALYSE THE COUNR OF APPROVED APPLICATIONS FROM MERGED DATA
plt.figure(figsize = (16, 6))
plt.bar(merged_appl_emp_ctgy.index.astype(str), 100 * merged_appl_emp_ctgy['TARGET'])

plt.xticks(rotation=90)
plt.xlabel('MERGED DATA - EMP EXPERIENCE FOR APPROVED APPLICATIONS WITH TARGET 1 (DEFAULT)')
plt.ylabel('DEFAULT %')
plt.title('MERGED DATA -EMPLOYMENT EXPERIENCE FOR APPROVED APPLICATIONS BI-VARIATE WITH TARGET DEFAULT')
plt.show()
```

```python
##PLOT THE EMPLOYMENT EXPERIENCE PLOT FOR CURRENT APPLICATION DATA
appl_emp = appl_df[['TARGET', 'DAYS_EMPLOYED']]
appl_emp['EMP_YEARS'] = appl_emp['DAYS_EMPLOYED'] / 365
appl_emp['EMP_YEARS_BINS'] = pd.cut(appl_emp['EMP_YEARS'], bins = np.linspace( 0.00, 50.00, num = 20))
appl_emp_ctgy  = appl_emp.groupby('EMP_YEARS_BINS').mean()
print(appl_emp_ctgy)
```

```python
##PLOT THE HISOGRAM WITH EMP EXPERIENCE BINS AND ANALYSE THE COUNT FROM ALL CURRENT DATA
plt.figure(figsize = (16, 6))
plt.bar(appl_emp_ctgy.index.astype(str), 100 * appl_emp_ctgy['TARGET'])

plt.xticks(rotation=90)
plt.xlabel('EMPLOYMENT EXPERIENCE WITH TARGET 1 (DEFAULT)')
plt.ylabel('DEFAULT %')
plt.title('EMPLOYMENT EXPERIENCE BI-VARIATE WITH TARGET DEFAULT')
plt.show()
```

## <font color=blue> From the Employment Experience Bi-variate plot for TARGET column, on both the CURRENT APPLICATION DATA & the MERGED DATA (OF APPROVED APPLICATIONS), it is clear that with high years of experience, there is a very less (almost to ZERO) chance of customer doing a DEFAULT on Loan </font>
